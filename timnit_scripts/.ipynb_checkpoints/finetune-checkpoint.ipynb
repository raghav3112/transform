```python
import os
import sys
sys.path.insert(0, '/scr/r6/tgebru/inverting_conv/caffe_invert_alexnet/python/')

import caffe
import numpy as np
from pylab import *
%matplotlib inline
```

```python
niter = 200
# losses will also be stored in the log
train_loss = np.zeros(niter)

#caffe.set_device(0)
#caffe.set_mode_gpu()
# We create a solver that fine-tunes from a previously trained network.
solver = caffe.SGDSolver('./solver.prototxt')
#solver.net.copy_from('invert_alexnet_conv1.caffemodel')
```

```python
# We run the solver for niter times, and record the training loss.
for it in range(niter):
    solver.step(1)  # SGD by Caffe
    # store the train loss
    train_loss[it] = solver.net.blobs['loss'].data
    if it % 10 == 0:
        print 'iter %d, finetune_loss=%f, scratch_loss=%f' % (it, train_loss[it], scratch_train_loss[it])
print 'done'
```
