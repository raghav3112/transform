```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
    
caffe_root= '/scr/r6/tgebru/timnit_caffe/'
import sys
sys.path.insert(0, caffe_root + 'python')

import caffe

plt.rcParams['figure.figsize'] = (50, 50)
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

import os
import copy
```

```python
caffe.set_mode_cpu()
net = caffe.Net(caffe_root + 'models/bvlc_reference_caffenet/deploy.prototxt',
                caffe_root + 'models/bvlc_reference_caffenet/bvlc_reference_caffenet.caffemodel',
                caffe.TEST)

# input preprocessing: 'data' is the name of the input blob == net.inputs[0]
transformer = caffe.io.Transformer({'data': net.blobs['data'].data.shape})
transformer.set_transpose('data', (2,0,1))
transformer.set_mean('data', np.load(caffe_root + 'python/caffe/imagenet/ilsvrc_2012_mean.npy').mean(1).mean(1)) # mean pixel
transformer.set_raw_scale('data', 255)  # the reference model operates on images in [0,255] range instead of [0,1]
transformer.set_channel_swap('data', (2,1,0))  # the reference model has channels in BGR order instead of RGB
```

```python
# set net to batch size of 20
source_ims='/imagenetdb3/tgebru/timelapse_experiments/tiny_source_unwarped'
target_ims='/imagenetdb3/tgebru/timelapse_experiments/tiny_target_unwarped' 

source_files=sorted(os.listdir(source_ims))
target_files=sorted(os.listdir(target_ims))
net.blobs['data'].reshape(20,3,227,227)

print source_files
print target_files

net_i=0
for s in source_files:
    net.blobs['data'].data[net_i,:,:,:] =transformer.preprocess('data', caffe.io.load_image(os.path.join(source_ims,s)))
    net_i += 1
      
for s in target_files:
    net.blobs['data'].data[net_i,:,:,:] =transformer.preprocess('data', caffe.io.load_image(os.path.join(target_ims,s)))
    net_i += 1
   
net.forward()
```

```python
#Plot TSNE visualization

from time import time
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import NullFormatter
from sklearn import manifold, datasets

base_dir='/imagenetdb3/tgebru/timelapse_experiments/tsne'
Orders=['C','F','A']
feats=[k for k in net.params]


#have same colors for same spot or for same weather.
color_names=['GPS','seasons']
color_season=np.concatenate((np.zeros(10), np.ones(10)),axis=0)
color_gps=np.zeros(20)
colors=np.random.rand(10)
c_i=0

for c in colors:
    color_gps[c_i]=c
    color_gps[c_i+10]=c
    c_i += 1
```

```python
#Loop and plot all conv features and try all order 
#for reshaping feature matrix
    
for feat in net.params:
    for Order in Orders:
        for color_name in color_names:
            if color_name=='seasons':
                color=color_season
            else:
                color=color_gps
            t0 = time()
            tsne = manifold.TSNE(n_components=2, init='pca', random_state=0)
            try:
                feat_shape=net.blobs[feat].data.shape[1]* \
                    net.blobs[feat].data.shape[2]*net.blobs[feat].data.shape[3]
            except:
                feat_shape=net.blobs[feat].data.shape[1]
            X=net.blobs[feat].data.reshape(20,feat_shape,order=Order)
            Y = tsne.fit_transform(X)
            t1 = time()
            print("t-SNE: %.2g sec" % (t1 - t0))
            fig = plt.figure(figsize=(15, 8))
            ax = fig.add_subplot(250)
            plt.scatter(Y[:, 0], Y[:, 1], c=color, cmap=plt.cm.Spectral)
            plt.title("t-SNE %s Order=%s color=%s(%.2g sec)" % \
                      (feat, Order, color_name,(t1 - t0)))
            ax.xaxis.set_major_formatter(NullFormatter())
            ax.yaxis.set_major_formatter(NullFormatter())
            plt.axis('tight')
            plt.show()
            img_file='%s/%s_%s_%s'%(base_dir,feat,Order,color_name)
            fig.savefig(img_file, bbox_inches='tight')
```

```python
#Represent each image as a list of matricies of the 
#inner product of its features
feat_mat=[]
num_data_points=net.blobs['conv1'].data.shape[0]
for data_i in xrange(0,num_data_points):
    data_i_feat={}
    print 'processing data %d out of %d'%(data_i,num_data_points)
    for feat in feats:
        gram_mat=np.zeros((net.blobs[feat].data[0].shape[0],
                           net.blobs[feat].data[0].shape[0]))
        for g in xrange(0,net.blobs[feat].data[0].shape[0]):
            for h in xrange(0,net.blobs[feat].data[0].shape[0]):
                if len(net.blobs[feat].data[0].shape) > 1:
                    f1=net.blobs[feat].data[data_i,g,:,:]
                    f2=net.blobs[feat].data[data_i,h,:,:]
                    gram_mat[g,h]=np.inner(f1.reshape((f1.shape[0]*f1.shape[1])),
                        f2.reshape((f2.shape[0]*f2.shape[1])))
        data_i_feat[feat]=gram_mat
    feat_mat.append(data_i_feat)
```
